# メインブロック

---




```python
#@title 変換器

TYPE = 'wav' #@param ["mp3", "wav"] {allow-input: false}
URL = "https://www.youtube.com/watch?v=xxxxxx" #@param {type:"string"}

import youtube_dl
from google.colab import files

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl':'%(title)s.' + TYPE,
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': TYPE,
        'preferredquality': '192',
    }],
}
with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    res = ydl.extract_info(URL, download=True)
    files.download("/content/" + res['title'] + "." + TYPE) 
```
